## Importando bibliotecas

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import matplotlib.pyplot as plt 
```

### Carregando dados e exibindo a descrição da base de dados

```python
california_housing = fetch_california_housing(as_frame=True)
print(california_housing['DESCR'])
```

#### Carrega os dados para um df pandas

```python
california_df = california_housing['data']
california_df
```

```python
california_df.shape
```

```python
california_df.head()
```

## Análise das variáveis

Para iniciar a análise das variáveis vamos visulizar seus tipos e quantidade de valores não nulos

```python
raw_california_df_copy = california_df.copy()
california_df.info()
```

Todas as variáveis são numéricas e continuas, então só precisamos nos preocupar em analisar esse tipo de variável.

```python
california_df.dtypes.value_counts()
```

Entendendo o universo de valores das variáveis

```python
summary = california_df \
    .select_dtypes('number') \
    .describe() \
    .transpose() \
    .sort_values(by='count', ascending=True)

summary
```

### Plotando as variáveis continuas

```python
for col, val in summary.items():
    plt.figure()
    summary[col].hist()
    plt.title(f'{col}')
    plt.show()
```
